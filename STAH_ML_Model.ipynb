now lets make a ensemble model using CatBoosting and RandomForest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X,y)

catboost_classifier = CatBoostClassifier(iterations=100, learning_rate=0.1, verbose=False)
catboost_classifier.fit(X,y)
```

```python
 import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier

def get_user_input(training_columns):
    gender = float(input("Enter gender: "))
    Age = int(input("Enter your age: "))
    year = int(input("Enter the year: "))
    cgpa = float(input("Enter your CGPA: "))
    anxiety = int(input("Do you have anxiety? (1 for Yes, 0 for No): "))
    course = str(input("Enter your course: "))

    new_data = pd.DataFrame(columns=training_columns)

    new_data.loc[0, 'gender'] = gender
    new_data.loc[0, 'Age'] = Age
    new_data.loc[0, 'year'] = year
    new_data.loc[0, 'cgpa'] = cgpa
    new_data.loc[0, 'anxiety'] = anxiety

    new_data[f'course_{course}'] = 1 if f'course_{course}' in new_data.columns else 0

    new_data = new_data[training_columns]

    new_data_filled = new_data.fillna(0)

    return new_data_filled

def ensemble_prediction(input_data):
    rf_proba = rf_classifier.predict_proba(input_data)[0]
    catboost_proba = catboost_classifier.predict_proba(input_data)[0]

    combined_proba = (rf_proba + catboost_proba) / 2
    predicted_class = combined_proba.argmax()

    return predicted_class

user_data = get_user_input(X.columns)  

predicted_class = ensemble_prediction(user_data)

if predicted_class == 1:
    print("The predicted depression for the user has been diagnosed.")
else:
    print("The predicted depression for the user has not been diagnosed.")
```
